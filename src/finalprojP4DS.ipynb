#### ***Question 1***: In which type of travel do the customer types have a higher ratio?<a id="41"></a>

- The benefits of  finding  the  answer
    - To determine which type of travel results in a higher customer retention rate. This will provide the airline with valuable insights into customer preferences.
- How to preprocess data:
    - We extract the columns, `Customer Type` and `Type of Travel`, as these are the only ones needed for this analysis.
- How to analyze data to answer the question:
    - Grouping data: Group the data by `Type of Travel` and `Customer Type`. This will help us to count the number of each customer type within each travel type.
    - Calculating Ratios: For each type of travel, we calculate the ratio of customer types. This involves summing the number of each customer type and then dividing by the total number of customers in that travel type.
    - Comparison: Compare these ratios to determine in which type of travel each customer type is more prevalent.

__data preprocessing__

```python
df_copy = airline_df.copy()

# count the number of each customer type
grouped_data = df_copy.groupby(['Type of Travel', 'Customer Type']).size().unstack()

# calculate ratio
ratios = grouped_data.div(grouped_data.sum(axis=1), axis=0)
```

__visualization__

```python
ratios.plot(kind='bar', stacked=True)
plt.title('Ratio of Customer Types in Different Types of Travel')
plt.xlabel('Type of Travel')
plt.ylabel('Ratio')
plt.xticks(rotation=0)
plt.legend(title='Customer Type')
plt.grid(axis='y', linestyle='-', alpha=0.5, color='black')
plt.gca().set_axisbelow(True)
plt.show()
```

__observation__

In this graph, we can easily see:

- For `Business travel`, there is a significant proportion of `Loyal Customers` (approximately 73.71%), with `disloyal Customers` making up about 26.29%.

- `Personal Travel` shows a very high ratio of `Loyal Customers`, almost reaching near exclusivity at 99.50%, with `disloyal Customers` at a mere 0.50%.

> From these observations, we can conclude that `Loyal Customers` constitute the majority in both travel types. However, the ratio of `Loyal Customers` is notably higher in `Personal Travel` compared to `Business travel`. This might suggest that personal travelers are more likely to stick with the same airline, indicating a potential area of focus for customer loyalty programs. Business travelers, while still primarily loyal, show a relatively higher presence of disloyal customers, which could indicate a competitive market or differing expectations not consistently met by the airline`s services

#### ***Question 2***: What is the distribution of customer satisfaction levels across different age groups, and do certain age groups tend to be more satisfied than others?<a id="42"></a>

- The benefits of  finding  the  answer is: 
    - Understanding the distribution of customer satisfaction across different age groups can help the airline to identify specific needs and preferences of each age segment.
    - This insight allows for targeted marketing strategies and service enhancements to improve overall customer satisfaction.
    - By recognizing which age groups are more or less satisfied, the airline can allocate resources more effectively to address any service gaps and enhance the customer experience.
    - This can lead to increased customer loyalty, repeat business, and positive word-of-mouth, which are critical factors for the airline`s success in a competitive market.
- How to preprocess data:
    - Clean the dataset by checking for missing values, especially in the `Age` and `satisfaction` columns, and decide on a strategy to handle them, like imputation or removal.
    - Convert the `Age` variable into categorical age groups to facilitate group-wise analysis.
    - Ensure that the `satisfaction` variable is in an appropriate format for analysis, potentially mapping it to an ordinal scale if it is categorical.
- How to analyze data to answer the question:
    - Group the data by the newly created age group categories and calculate the distribution of satisfaction levels within each group.
    - Use visualizations, such as stacked bar charts or box plots, to illustrate the distribution of satisfaction levels across age groups.
    - Perform statistical tests if necessary to determine whether the differences in satisfaction levels between age groups are significant.
    - Interpret the results to determine if certain age groups are more satisfied than others, and provide recommendations based on these findings.

__data preprocessing__

```python
age_bins = [17, 30, 40, 50, 60, 70, 80]
age_labels = ['18-30', '31-40', '41-50', '51-60', '61-70', '71+']

df_copy['Age_Group'] = pd.cut(df_copy['Age'], bins=age_bins, labels=age_labels, right=True)

# Check the distribution of satisfaction across age groups
age_satisfaction_distribution = df_copy.groupby(['Age_Group', 'satisfaction'], observed=True).size().unstack(fill_value=0)

# Normalize the distribution to get proportions instead of counts
age_satisfaction_distribution = age_satisfaction_distribution.divide(age_satisfaction_distribution.sum(axis=1), axis=0)
```

__visualization__

```python
age_satisfaction_distribution.plot(kind='bar', stacked=True, colormap='plasma', alpha=0.85)
plt.title('Distribution of Customer Satisfaction Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Proportion of Satisfaction')
plt.legend(title='Satisfaction', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

__observation__

- Satisfaction levels appear relatively consistent across all age groups, with a larger share of customers reporting satisfaction in each category.

- Younger (18-30) and older (71+) age groups show a slightly higher proportion of satisfaction compared to those in the middle age ranges.

- No single age group stands out as significantly more satisfied or dissatisfied, suggesting that age might not be a sole predictor of satisfaction.

- Both the youngest and the oldest customers tend to report higher satisfaction, indicating potentially different expectations or experiences that are better met by the airline's services for these age groups.

#### ***Question 3***: Does the length of arrival delay affect the overall satisfaction differently for business and personal travelers?<a id="43"></a>

- Benefits of  finding  the  answer: 
    -  Understanding how arrival delays impact satisfaction for different traveler segments allows airlines and travel service providers to tailor their customer service strategies.
    - Identifying the factors contributing to dissatisfaction due to arrival delays can help airlines optimize their operations. This may involve refining scheduling, communication, or logistics to minimize delays, 
- How to preprocess data:
    - We extract the columns, including `satisfaction`, `Type of Travel`, and `Arrival Delay in Minutes`, as these are the only ones needed for this analysis.
    - We check missing values and preprocessing these values by removing if needed. 
    - We identify ranges which is not outliers
- How to analyze data to answer the question:
    - We will draw 2 chart representing `Personal Travel` and `Business travel` to access the distribution of `Arrival Delay in Minutes` of the satisfied and dissatisfied people.


__data preprocessing__

```python
# Preparing Data
df = airline_df[['satisfaction','Type of Travel','Arrival Delay in Minutes']]
df.isna().mean(axis=0) * 100
```

```python
# Cleaning data
df = df.dropna()
```

```python
# Calculate Q1, Q3, and IQR
Q1 = df['Arrival Delay in Minutes'].quantile(0.25)
Q3 = df['Arrival Delay in Minutes'].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds for potential outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify ranges
range_df = df[(df['Arrival Delay in Minutes'] >= lower_bound) & (df['Arrival Delay in Minutes'] <= upper_bound)]
```

__visualization__

```python
travel_type = range_df['Type of Travel'].unique().tolist()
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,8), sharey=True)

for index, val in enumerate(travel_type):
    Travel_df = range_df[range_df['Type of Travel'] == travel_type[index]]
    satisfied_df = Travel_df.loc[Travel_df ['satisfaction'] == 'satisfied']
    dissatisfied_df = Travel_df.loc[Travel_df ['satisfaction'] == 'dissatisfied']
    sns.kdeplot(data=satisfied_df['Arrival Delay in Minutes'], label='satisfied',fill=True,ax=axes[index])
    sns.kdeplot(data=dissatisfied_df['Arrival Delay in Minutes'], label='dissatisfied',fill=True,ax=axes[index])
    axes[index].legend(title='Satisfaction', loc='upper right')
    axes[index].set_xlabel('Arrival Delay in Minutes')
    axes[index].set_ylabel('Density')
    axes[index].set_title(f'{val}')

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

__observation__
- The two above charts are charts showing the distribution of flight delay times in minutes. Depending on different travel purposes, there can be slight differences. The distribution of arrival delays in minutes by types of travel and customer satisfaction is noticeable for being right-skewed as observed in the density plot. This indicates that the Arrival Delay in minutes does not variously affect customer satisfaction.

- In general, we see that the probability density of the satisfied and dissatisfied people is not high. The probability density of satisfied customers is higher than dissatisfied customers in both the charts. The charts also show significant variation in customer satisfaction at all times. In particular, the arrival delay in minutes is 0, and the probability density of satisfied and dissatisfied people is the highest. The possible explanation for this variation could be that during the flight, other services satisfied many people and also made many other people dissatisfied.
- Arrival delays are sometimes unpredictable and inevitable, so we can use services to satisfy customers. However, an arrival delay is also an indispensable factor that can cause customer disappointment.

#### ***Question 4***: How does the satisfaction rate change according to customer age?<a id="44"></a>

- The benefits of  finding  the  answer: 
    - Understanding preferences and expectations of different age groups can help tailor services to meet their specific needs.
- How to preprocess data:
    - We extract the columns, `Age` and `satisfaction`, as these are the only ones needed for this analysis.
    - Convert the `satisfaction` column into a numerical format for correlation analysis (e.g., 1 for satisfied, 0 for not satisfied).
- How to analyze data to answer the question:
    - Grouping and Averaging: Group the data by `Age` and calculate the satisfaction ratio by calculating the average satisfaction score for each age group. This provides an idea of how satisfaction changes with age.
    - Visualization: Plot the average satisfaction score against age to visually assess any trends or patterns.

__data proprocessing__

```python
df_copy = airline_df.copy()
df_copy = df_copy[['Age', 'satisfaction']]

# assign 1 for satisfied, 0 for not satisfied
df_copy['satisfaction_binary'] = df_copy['satisfaction'].apply(lambda x: 1 if x == 'satisfied' else 0)

# calculate mean
age_satisfaction = df_copy.groupby('Age')['satisfaction_binary'].mean()
```

__visualization__

```python
plt.figure(figsize=(10, 6))
age_satisfaction.plot()
plt.title('Satisfaction rate by age')
plt.xlabel('Age')
plt.ylabel('Satisfaction rate (1 = Satisfied, 0 = Not Satisfied)')
plt.grid(True)
plt.show()
```

__observation__
- The chart shows that the average customer satisfaction rate for airlines is highest for customers in from 40s to 60s (is greater than 0.6). We can figure out that people from the ages of 40 to 60 are very satisfied with airline services. This rate then declines for younger and older customers. 
- For older customers, the satisfaction rate is less than 0.5 and may reach the lowest rate. We can explain this problem following: 
    - They prioritize comfort and have heightened concerns about health during travel,so they may dissatisfied relevant services. 
    - Older customers may face challenges adapting to modern technology, and if the airline services are not user-friendly or accommodating, it might contribute to dissatisfaction.
- For younger customers, the satisfaction rate is not high and less than 0.55.
    - They may prioritize technology-related services such as inflight WiFi, online support ,online booking and on-board service. If these services are not up to par, it could impact satisfaction.
- Besides, there are other factors, which can be Food and drink, Entertainment, and so on, lead to customer dissatisfaction
